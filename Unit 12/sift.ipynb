```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv.imread(r'D:\Cursos_Online\computer-vision-beginners\Unit 12\1.jpg')
img2 = cv.imread(r'D:\Cursos_Online\computer-vision-beginners\Unit 12\2.jpg')
gray1 = cv.cvtColor(img1, cv.COLOR_BGR2GRAY)
gray2 = cv.cvtColor(img2, cv.COLOR_BGR2GRAY)
sift = cv.SIFT_create()
kp1 = sift.detect(gray1, None)
kp2 = sift.detect(gray2, None)
img1 = cv.drawKeypoints(gray1, kp1, img1, flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
img2 = cv.drawKeypoints(gray2, kp2, img2, flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.subplot(121)
plt.imshow(img1[:,:,::-1])
plt.subplot(122)
plt.imshow(img2[:,:,::-1])
```

```python
img1 = cv.imread(r'D:\Cursos_Online\computer-vision-beginners\Unit 12\1.jpg')
img2 = cv.imread(r'D:\Cursos_Online\computer-vision-beginners\Unit 12\2.jpg')
sift = cv.SIFT_create()
kp1, des1 = sift.detectAndCompute(gray1, None)
kp2, des2 = sift.detectAndCompute(gray2, None)
```

```python
bf = cv.BFMatcher(cv.NORM_L2, crossCheck=True)
matches = bf.match(des1, des2)
```

```python
matches = sorted(matches, key = lambda x:x.distance)
```

```python
img3 = cv.drawMatches(img1, kp1, img2, kp2, matches[:20], img2, flags=2)
```

```python
plt.figure(figsize=(15,20))
plt.imshow(img3[:,:,::-1])
```
