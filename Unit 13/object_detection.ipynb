```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
from imutils.object_detection import non_max_suppression
```

```python
imPath = r'D:\Cursos_Online\computer-vision-beginners\Unit 13\p.jpg'
im = cv.imread(imPath)
```

```python
plt.imshow(im[:,:,::-1])
```

```python
hog = cv.HOGDescriptor()
hog.setSVMDetector(cv.HOGDescriptor_getDefaultPeopleDetector())
```

```python
(r, w) = hog.detectMultiScale(im, winStride=(4,4), padding=(16,16), scale=(2))
```

```python
rects = np.array([[x,y,x+w,y+h] for(x,y,w,h) in r])
```

```python
rects_NMS = non_max_suppression(rects, probs=None, overlapThresh=0.6)
```

```python
for (x1, y1, x2, y2) in rects_NMS:
    cv.rectangle(im, (x1, y1), (x2, y2), (0,0,255), 2)
```
